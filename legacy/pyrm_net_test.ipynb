```python
from pyrm_net import *
from keras.utils.np_utils import to_categorical 
from keras.datasets import cifar10
import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
print(x_train.shape)
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)
print(y_train.shape)
```

```python
model = build_pyrm_net(3, 10, max_pool_alt=True)
model.summary()
model.fit(x_train,y_train, epochs = 3, verbose = 1, validation_split= 0.02)
```

```python
model = build_pyrm_net(3, 10, max_pool_alt=False)
model.summary()
model.fit(x_train,y_train, epochs = 3, verbose = 1, validation_split= 0.02)
```
